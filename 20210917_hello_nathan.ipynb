## Opgave

Opdracht
Schrijf een programma dat de gebruiker om zijn/haar voornaam vraagt.

Daarna begroet het programma de gebruiker door “Hallo, voornaam!” op het scherm te tonen.

Dus:

* Hallo
* een komma
* een spatie
* de voornaam van de gebruiker
* een uitroepteken


Het programma …

* vraagt de gebruiker naar een (voor)naam en kent deze waarde toe aan een variabele (bv. voornaam).
* stelt een nieuwe ‘string’ samen (beschrijving hierboven).
* drukt de nieuwe string af op het scherm.
