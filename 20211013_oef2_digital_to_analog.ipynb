# Achtergrondinformatie

De digitale klok geeft aan dat het 18:20:00 is. Dit komt overeen met de aanduiding 20 over 6 op de analoge klok.

# Opgave

- Schrijf een programma dat op basis van het digitale tijdstip de positie van de kleine wijzer aangeeft als hoek in graden tov de 12u positie.

- We geven van de digitale klok enkel de uren en minuten in. We gaan er vanuit dat de seconden op 0 staan.

# Een mogelijke aanpak
Het programma 

- vraagt de gebruiker naar de uren, en kent deze waarde toe aan een variabele. (bv. uren, u, )
- vraagt de gebruiker naar de minuten (enkel even waarden!), en kent deze waarde toe aan een variabele. (bv. minuten, min, m, )
- bepaalt de daaruitvolgende positie van de kleine wijzer (eveneens uitgedrukt in graden tov de 12u positie).
- drukt de gevonden waarde af.
